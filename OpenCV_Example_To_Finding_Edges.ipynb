Below, you've been given pne common type of edge detection filter: a Sobel Opeartor

The Sobel Opearato is very commonly used in edge detection and in finding patterns in intensity in an image. Applying a Sobel Filter to an image is a way taking (an approximation) of the derivative of the image in the x or y direction, separately.

```python
# Create a custom kernel
# 3x3 array for edge detection
sobel_y = np.array([[-1, -2, -1],
                   [0, 0, 0],
                   [1, 2, 1 ]])

# TODO: Create and apply a Sobel x Operator

# Filter the image using filter2D, which has inputs: (grayscale image, bit-depth, kernel)
filtered_image = cv2.filter2D(gray, -1, sobel_y)

plt.imshow(filtered_image, cmap='gray')
```
